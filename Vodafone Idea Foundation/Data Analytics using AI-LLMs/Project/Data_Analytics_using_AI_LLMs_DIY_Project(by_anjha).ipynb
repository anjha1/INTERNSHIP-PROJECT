```python
print(data[['Sales', 'Profit']].corr())
```

```python
category_analysis = data.groupby(['Category', 'Sub-Category'])[['Sales', 'Profit']].sum()
print(category_analysis)
```

```python

category_sales_profit = data.groupby('Category')[['Sales', 'Profit']].sum().reset_index()

# Barplot for Sales and Profit by Category
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
sns.barplot(data=category_sales_profit, x='Category', y='Sales', ax=ax[0])
ax[0].set_title('Total Sales by Category')
sns.barplot(data=category_sales_profit, x='Category', y='Profit', ax=ax[1])
ax[1].set_title('Total Profit by Category')
plt.show()
```

```python
# Total Sales and Profit by Region
region_sales_profit = data.groupby('Region')[['Sales', 'Profit']].sum().reset_index()

# Barplot for Sales and Profit by Region
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
sns.barplot(data=region_sales_profit, x='Region', y='Sales', ax=ax[0])
ax[0].set_title('Total Sales by Region')
sns.barplot(data=region_sales_profit, x='Region', y='Profit', ax=ax[1])
ax[1].set_title('Total Profit by Region')
plt.show()
```

```python
# Sales Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Sales'], bins=10, kde=True)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# 3D Scatter plot of Sales, Profit, and Quantity
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(data['Sales'], data['Profit'], data['Quantity'], c='blue', marker='o')

# Adding axis labels
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
ax.set_zlabel('Quantity')
ax.set_title('3D Scatter Plot of Sales, Profit, and Quantity')
plt.show()
```

```python
# Correlation Matrix Heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = data[['Sales', 'Profit', 'Quantity', 'Discount']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Pair plot to explore pairwise relationships
sns.pairplot(data[['Sales', 'Profit', 'Quantity', 'Discount']])
plt.show()
```

```python
# Scatter plot of Quantity vs Sales and Quantity vs Profit
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
sns.scatterplot(data=data, x='Quantity', y='Sales', ax=ax[0], color='blue')
ax[0].set_title('Quantity vs Sales')
sns.scatterplot(data=data, x='Quantity', y='Profit', ax=ax[1], color='green')
ax[1].set_title('Quantity vs Profit')
plt.show()
```

```python
# Sales and Profit by Ship Mode
ship_mode_sales_profit = data.groupby('Ship Mode')[['Sales', 'Profit']].sum().reset_index()

# Barplot for Sales and Profit by Ship Mode
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
sns.barplot(data=ship_mode_sales_profit, x='Ship Mode', y='Sales', ax=ax[0], palette="Blues")
ax[0].set_title('Total Sales by Ship Mode')
sns.barplot(data=ship_mode_sales_profit, x='Ship Mode', y='Profit', ax=ax[1], palette="Greens")
ax[1].set_title('Total Profit by Ship Mode')
plt.show()
```

```python
# Sales and Profit by Segment
segment_sales_profit = data.groupby('Segment')[['Sales', 'Profit']].sum().reset_index()

# Barplot for Sales and Profit by Segment
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
sns.barplot(data=segment_sales_profit, x='Segment', y='Sales', ax=ax[0], palette="Purples")
ax[0].set_title('Total Sales by Segment')
sns.barplot(data=segment_sales_profit, x='Segment', y='Profit', ax=ax[1], palette="Oranges")
ax[1].set_title('Total Profit by Segment')
plt.show()
```

```python
# Sales and Profit by City
city_sales_profit = data.groupby('City')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False).head(10).reset_index()

# Barplot for Sales and Profit by City
fig, ax = plt.subplots(1, 2, figsize=(14, 8))
sns.barplot(data=city_sales_profit, x='Sales', y='City', ax=ax[0], palette="viridis")
ax[0].set_title('Top 10 Cities by Sales')
sns.barplot(data=city_sales_profit, x='Profit', y='City', ax=ax[1], palette="magma")
ax[1].set_title('Top 10 Cities by Profit')
plt.show()
```

```python
# Scatter plot of Discount vs Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Discount', y='Profit', color='coral')
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

```python
# Profit Margin by State
state_profit_margin = data.groupby('State')['Profit Margin (%)'].mean().reset_index().sort_values(by='Profit Margin (%)', ascending=False)

# Barplot for Profit Margin by State (top 10 states)
plt.figure(figsize=(10, 6))
sns.barplot(data=state_profit_margin.head(10), x='Profit Margin (%)', y='State', palette="Blues")
plt.title('Top 10 States by Profit Margin')
plt.show()
```

```python
# Top Products by Sales
top_sales = data.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()
top_profit = data.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).head(10).reset_index()

# Plotting Sales and Profit for Top 10 Products
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
sns.barplot(data=top_sales, x='Sales', y='Sub-Category', ax=ax[0], palette="viridis")
ax[0].set_title('Top 10 Products by Sales')
sns.barplot(data=top_profit, x='Profit', y='Sub-Category', ax=ax[1], palette="magma")
ax[1].set_title('Top 10 Products by Profit')
plt.show()
```

```python
# Preparing data for clustered bar chart
region_category = data.groupby(['Region', 'Category'])[['Sales', 'Profit', 'Quantity']].sum().reset_index()

# Clustered Bar Chart
fig, ax = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
sns.barplot(data=region_category, x='Region', y='Sales', hue='Category', ax=ax[0], palette="muted")
ax[0].set_title('Sales by Region and Category')
sns.barplot(data=region_category, x='Region', y='Profit', hue='Category', ax=ax[1], palette="pastel")
ax[1].set_title('Profit by Region and Category')
sns.barplot(data=region_category, x='Region', y='Quantity', hue='Category', ax=ax[2], palette="bright")
ax[2].set_title('Quantity by Region and Category')
plt.show()
```
